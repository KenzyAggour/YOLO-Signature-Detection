```python
from ultralytics import YOLO
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
model = YOLO('yolov8m.yaml')
```

```python
model.train(data = '/kaggle/input/yaml-formated-file/config.yaml' , 
            epochs = 20 , seed = 3 , batch = 5)
```

```python
def detect(img_path):
    img = cv2.imread(img_path)
    det = model(img)
    det = det[0].plot()
    return det

res = detect('/kaggle/input/signatures/SignatureObjectDetection/Train/images/10.tif')
plt.imshow(res)
```

```python
res = detect('/kaggle/input/signatures/SignatureObjectDetection/Test/images/682.tif')
plt.imshow(res)
```
